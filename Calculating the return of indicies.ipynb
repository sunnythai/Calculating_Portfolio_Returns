```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC','^IXIC','^GDAXI','^N225'] #list
ind_data=pd.DataFrame()
for t in tickers: #for loop 
    ind_data[t]=wb.DataReader(t, data_source='yahoo', start='2000-1-4')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail
```

```python
(ind_data/ind_data.iloc[0]*100).plot(figsize=(25,13));
plt.show()
```

```python
ind_returns=(ind_data/ind_data.shift(1))-1
ind_returns.tail()
```

```python
annual_ind_returns=ind_returns.mean()*250
annual_ind_returns
```
