```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers= ["TSLA", "CRSP", "ROKU", "TDOC", "SQ",
          "NVTA", "TWST","PACB", "PSTG", "Z",
          "EDIT", "IOVA", "SPOT", "NTLA","EXAS",
          "PRLB","TWLO", "MTLS","PYPL","BIDU", 
          "SHOP", "DOCU","ICE", "TSM", "TREE"]

print (len(tickers))
mydata=pd.DataFrame()
for t in tickers:
    mydata[t]=wb.DataReader (t, data_source='yahoo', start='2020-1-1', end='2020-12-24')['Adj Close']
```
